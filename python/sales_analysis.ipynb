```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing files to JupiterNotebook
data_year = pd.read_csv("sales_per_year.csv")
data_month = pd.read_csv("sales_per_month.csv")
data_quarter = pd.read_csv("sales_per_quarter.csv")
```

```python
#Creating Visual grahp to show total sales per reported years

data_year.groupby("year")["total_sales"].sum().plot(kind="bar")
plt.title ("Sales per Year")
plt.xlabel ("Year")
plt.ylabel ("Total Sales")
plt.show()
```

```python
#Visual grahp to show total sales reported year quarters
data_quarter["year_quarter"] = data_quarter["year"].astype(str) + " Q" + data_quarter["quarter"].astype(str)

plt.figure(figsize=(10,5))
plt.bar(data_quarter["year_quarter"], data_quarter["total_sales"], color="skyblue")
plt.title("Sales by quarters, year-on-year")
plt.xlabel("Year and Quarter")
plt.ylabel("Total Sales")
plt.xticks(rotation=15)
plt.show()
```
