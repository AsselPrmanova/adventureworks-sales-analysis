```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing files to JupiterNotebook
data_year = pd.read_csv("sales_per_year.csv")
data_month = pd.read_csv("sales_per_month.csv")
data_quarter = pd.read_csv("sales_per_quarter.csv")
```

```python
#Creating Visual grahp to show total sales per year

data_year.groupby("year")["total_sales"].sum().plot(kind="bar")
plt.title ("Sales per Year")
plt.xlabel ("Year")
plt.ylabel ("Total Sales")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18,5))  # 1 , 3 

#   
axes[0].bar(data_year["year"], data_year["total_sales"], color="skyblue")
axes[0].set_title("  ")
axes[0].set_xlabel("")
axes[0].set_ylabel(" ")

#   
axes[1].bar(data_quarter["quarter"], data_quarter["total_sales"], color="salmon")
axes[1].set_title("  ")
axes[1].set_xlabel("")

#   
axes[2].bar(data_month["month"], data_month["total_sales"], color="lightgreen")
axes[2].set_title("  ")
axes[2].set_xlabel("")

plt.tight_layout()  #     
plt.show()
```
